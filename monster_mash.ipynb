```python
class Player():
    def __init__(self, name, coords=[0,0], lives=3):
        self.name = name
        self.lives = lives
        self.coords = coords  #[Col, row]
        
    def setCoords(self, coords):
        self.coords = coords
        
    def getCoords(self):
        return self.coords
    def getName(self):
        self.name
    
    def decrementLives(self):
        self.lives -= 1
    
        
    def setLives(self, num):
        self.lives = num
    def getLives(self):
        return self.lives
        
    def movePlayer(self, move):
        if move == 'up' and self.coords[1] != 0:
            self.coords[1] -= 1
        elif move == 'down' and self.coords[1] != 4:
            self.coords[1] += 1
        elif move == 'left' and self.coords[0] != 0:
            self.coords[0] -= 1
        elif move == 'right' and self.coords[0] != 4:
            self.coords[0] += 1
        else:
            print('Invalid Response. Please try again.')
    
class Monster():
    def __init__(self,coords):
        self.coords = coords  #[Col, row]
        
    def getCoords(self):
        return self.coords
        
    def moveMonster(self, cols, rows):
        self.coords = [randint(0, cols - 1), randint(0, rows - 1)] # col , row
        
        
class Egg():
    #I want to create multiple eggs
    def __init__(self, coords):
        self.coords = coords
        
    def getCoords(self):
        return self.coords
    



class Door():
    def __init__(self, coords=[2,0]):
        
        self.coords = coords  #[Col, row]
        
    def getCoords(self):
        return self.coords
        
    



```
